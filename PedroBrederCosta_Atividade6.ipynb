## **Exercício 1**


```python
class Aluno():
  def __init__(self, nome, matricula, notas):
    self.nome = nome
    self.matricula = matricula
    self.notas = notas

  def calcular_media(self):
    return sum(self.notas)/len(self.notas)

  def situacao(self):
    if self.calcular_media() >= 6:
      return(f"Aprovado")
    else:
      return("Reprovado")

class AlunoBolsista(Aluno):
  def __init__(self, nome, matricula, notas, percentual_bolsa):
    super().__init__(nome, matricula, notas)
    self.percentual_bolsa = percentual_bolsa

  def manter_bolsa(self):
    if self.calcular_media() >= 7:
      return(f"Bolsa mantida")
    else:
      return("Bolsa cancelada")
```

```python
alunos = [
    Aluno('Pedro', 202465680, [10, 8.5, 9.65]),
    Aluno('Breder', 202467028, [9.7, 9.9, 10]),
    Aluno('Costa', 202665001, [9, 8.4, 9.85]),
    AlunoBolsista('Ruan', 202424024, [2, 4, 2.4], 20),
    AlunoBolsista('Guilerme', 202222022, [2.2, 0.2, 0], 10)
    ]

for aluno in alunos:
  if isinstance(aluno, AlunoBolsista):
    print(f"Aluno: {aluno.nome} // Matrícula: {aluno.matricula} // Média: {aluno.calcular_media():.2f} // Situação: {aluno.situacao()} // Percentual de Bolsa: {aluno.percentual_bolsa}% // Situação da Bolsa: {aluno.manter_bolsa()}")
  else:
    print(f"Aluno: {aluno.nome} // Matrícula: {aluno.matricula} // Média: {aluno.calcular_media():.2f} // Situação: {aluno.situacao()}")
```

## **Exercício 2**

```python
class Conta():
  def __init__(self, titular, numero):
    self.titular = titular
    self.numero = numero
    self._saldo = 0

  # uma boa prática ao lidar com manipulações de atributos privados é fazer
  # tais ações por meio de métodos, assim acessando os atributos dentro da
  # classe. Sendo assim,

  def depositar(self, valor):
    if valor > 0:
      self._saldo += valor
      print(f'Deposito feito. (+{valor})')
    else:
      print('Valor inválido para o depoósito.')

  def sacar(self, valor):
    if self._saldo - valor >= 0:
      self._saldo -= valor
      print(f'Saque efetudado. (-{valor})')
    else:
      print('Saque não efetudado.')

  def obter_saldo(self):
    return(self._saldo)

class ContaEspecial(Conta):
  def __init__(self, titular, numero, limite):
    super().__init__(titular, numero)
    self.limite = limite

  def sacar(self, valor):
    saldo_total = self._saldo + self.limite
    if valor <= saldo_total:
        self._saldo -= valor
        print(f'Saque efetudado. (-{valor})')
    else:
      print('Não foi possível efetuar o saque.')
```

```python
cc = Conta('Pedro', 10065)
ce = ContaEspecial('Breder', 10094, 1000) # limite definido como sendo 1000

deposito = 500
saque_menor = 250
saque_maior = 1000

print('Operações na conta comum:\n')
cc.depositar(deposito)
cc.sacar(saque_menor)
cc.sacar(saque_maior)

print('\nOperações na conta especial:\n')
ce.depositar(deposito)
ce.sacar(saque_menor)
ce.sacar(saque_maior)

print(f'\nSaldo da conta comum: {cc.obter_saldo():.2f}')
print(f'Saldo da conta especial: {ce.obter_saldo():.2f}')
```
