```python
class ContaBancaria:
    def __init__(self, titular, saldo = 0, limite = 1000):
        self.titular = titular
        self.saldo = saldo
        self.limite = limite

    def depositar(self, valor):
      self.saldo += valor
      print(f'Foram depositados R${valor:.2f}.')

    def sacar(self, valor):
      if valor <= self.saldo:
        self.saldo -= valor
        print('Saque efetuado com sucesso!')

      elif valor > self.saldo + 1000:
        if input('O valor de saque desejado é maior do que o saldo em conta somado ao seu limite, deseja fazer um empréstimo de até R$1000? (s/n)\n') == 's':
          while True:
            emprestimo = float(input("\nQuanto deseja pegar de EMPRÉSTIMO?\n"))

            if emprestimo > 1000 or emprestimo < 0:
              print(f'Por favor, digite um valor válido (entre R$0,00 e R${self.limite:.2f}):\n')
            else:
              break
        else:
          return 0

        self.saldo -= (self.saldo + emprestimo)
        print('Saque efetuado com sucesso!')

      else:
        if input('O valor de saque desejado é maior do que o saldo em conta, deseja fazer um empréstimo? (s/n)\n') == 's':
          self.saldo -= valor
          print('Saque efetuado com sucesso!')

    def exibir_saldo(self):
      print(f'\nEste é seu saldo atual: R${self.saldo:.2f}.')
```

```python
contas = [ContaBancaria('Pedro'), ContaBancaria('Breder')]

while True:
  opcao = input('\nMENU BANCÁRIO\n1. Escolher conta\n2. Sair\nDigite sua opção: ')

  if opcao == '1':
    indice = int(input('\nQual conta deseja acessar? (1 ou 2)\n'))
    conta = contas[indice-1]

    while True:

      opcao = input(f'\nMENU DE ESCOLHAS ({conta.titular})\n1. Depositar\n2. Sacar\n3. Mostrar saldo\n4. Voltar ao menu inicial\nDigite sua opção: ')

      if opcao == '1':
        deposito = float(input('\nQuanto deseja depositar?\n'))
        conta.depositar(deposito)

      elif opcao == '2':
        saque = float(input('\nQuanto deseja sacar?\n'))
        conta.sacar(saque)

      elif opcao == '3':
        conta.exibir_saldo()

      elif opcao == '4':
        break

      else:
        print('Insira uma opção válida.')

  elif opcao == '2':
    break

  else:
    print('Insira uma opção válida.')
```

### *Código usando métodos estáticos*

```python
class Conversor:

    @staticmethod
    def celsius_para_fahrenheit(c):
        return (c * 9/5) + 32

    @staticmethod
    def km_para_milhas(km):
        return km * 0.621371

    @staticmethod
    def reais_para_dolar(valor, cotacao):
        return valor / cotacao

while True:

  opcao = input('\nMENU\n1. Celsius para Fahrenheit\n2. Km para Milhas\n3. Reais para Dólar\n4. Sair\n')

  if opcao == '1':
    temp = float(input('\nQual a temperatura em Celsius que deseja converter para Fahrenheit?\n'))
    print(f'{temp}°C equivale a {Conversor.celsius_para_fahrenheit(temp):.2f}°F.')
  elif opcao == '2':
    dist = float(input('\nQual a distância em quilômetros que deseja converter para milhas?\n'))
    print(f'{dist} Km equivale a {Conversor.km_para_milhas(dist):.2f} milhas.')
  elif opcao == '3':
    valor = float(input('\nInsira o valor em R$ para a conversão para dólar: '))
    cot = float(input('Digite a cotação do dólar atual: '))
    print(f'R${valor:.2f} vale US${Conversor.reais_para_dolar(valor,cot):.2f}')
  elif opcao == '4':
    break
  else:
    print('Insira uma opção válida.')
```

### *Comentários e observações*

Note que, ainda que esteja dentro de uma classe, os métodos **não precisam do parâmetro `self`** que faz referência à própria classe. Além disso, nota-se que **não foi necessário criar um objeto** para que se utilizasse os métodos da classe. Dessa forma, é possível **assemelhar** os **métodos estáticos a funções comuns**. Note como o código a seguir e aquele executado anteriormente funcionam da mesma forma.

####*Código usando funções*

```python
def celsius_para_fahrenheit(c):
  return c * 9/5 + 32

def km_para_milhas(km):
  return km * 0.621371

def reais_para_dolar(valor, cotacao):
  return valor / cotacao

while True:

  opcao = input('\nMENU\n1. Celsius para Fahrenheit\n2. Km para Milhas\n3. Reais para Dólar\n4. Sair\n')

  if opcao == '1':
    temp = float(input('\nQual a temperatura em Celsius que deseja converter para Fahrenheit?\n'))
    print(f'{temp}°C equivale a {celsius_para_fahrenheit(temp):.2f}°F.')
  elif opcao == '2':
    dist = float(input('\nQual a distância em quilômetros que deseja converter para milhas?\n'))
    print(f'{dist} Km equivale a {km_para_milhas(dist):.2f} milhas.')
  elif opcao == '3':
    valor = float(input('\nInsira o valor em R$ para a conversão para dólar: '))
    cot = float(input('Digite a cotação do dólar atual: '))
    print(f'R${valor:.2f} vale US${reais_para_dolar(valor,cot):.2f}')
  elif opcao == '4':
    break
  else:
    print('Insira uma opção válida.')
```

####*Código usando métodos dinâmicos*

Por fim, note a diferença na escrita de um código que faz uso de métodos estáticos e outro que faz uso de métodos dinâmicos, executando códigos que fazem a mesma função, ainda que possuam essa diferença na sintaxe.

```python
class Conversor:

    def celsius_para_fahrenheit(self, c):
        return c * 9/5 + 32

    def km_para_milhas(self, km):
        return km * 0.621371

    def reais_para_dolar(self, valor, cotacao):
        return valor / cotacao

converte = Conversor() # por ser métodos dinâmicos, há necessidade de se criar um objeto

while True:

  opcao = input('\nMENU\n1. Celsius para Fahrenheit\n2. Km para Milhas\n3. Reais para Dólar\n4. Sair\n')

  if opcao == '1':
    temp = float(input('\nQual a temperatura em Celsius que deseja converter para Fahrenheit?\n'))
    print(f'{temp}°C equivale a {converte.celsius_para_fahrenheit(temp):.2f}°F.')
  elif opcao == '2':
    dist = float(input('\nQual a distância em quilômetros que deseja converter para milhas?\n'))
    print(f'{dist} Km equivale a {converte.km_para_milhas(dist):.2f} milhas.')
  elif opcao == '3':
    valor = float(input('\nInsira o valor em R$ para a conversão para dólar: '))
    cot = float(input('Digite a cotação do dólar atual: '))
    print(f'R${valor:.2f} vale US${converte.reais_para_dolar(valor,cot):.2f}')
  elif opcao == '4':
    break
  else:
    print('Insira uma opção válida.')
```

Observe como todos os três diferentes códigos tiveram a mesma saída para os mesmos *inputs*, isso confirma que todos são válidos, porém possuem formas diferentes de serem escritas.
